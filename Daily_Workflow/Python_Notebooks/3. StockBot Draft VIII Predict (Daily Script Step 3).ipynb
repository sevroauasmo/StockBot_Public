```python
loaded_data = loaded_data.dropna()
```

```python
num_and_dummies = loaded_data[['index', 'total_amount_bought_or_sold', 'total_dollar_bought_or_sold',
       'amount_owned_following', 'shares_percent_increased', 'dof_price',
       'vix_on_dof', 'tnx_on_dof', 'minus_10_price', '10_day_delta', 'purchase_price_vs_price_on_dof',
       '30_day_delta', 'minus_30_price', 'dof_vs_30_avg', 'dof_vs_30_actual']]
```

```python
final = pd.get_dummies(num_and_dummies).dropna()
```

```python
final
```

```python
# split data into features and target variable
X = final[['index', 'total_amount_bought_or_sold', 'total_dollar_bought_or_sold',
       'amount_owned_following', 'shares_percent_increased', 'dof_price',
       'vix_on_dof', 'tnx_on_dof', 'minus_10_price', '10_day_delta', 'purchase_price_vs_price_on_dof',
       '30_day_delta', 'minus_30_price']]

```

```python
X = loaded_scaler.transform(X.drop('index', axis=1))
```

```python
predictions = loaded_model.predict(X)
```

```python
predictions_1d = predictions.ravel()
```

```python
loaded_data['Predictions'] = predictions_1d
```

```python
loaded_data_sorted = loaded_data.sort_values(by='Predictions', ascending=False)
```

```python
#limit only to where the model thinks it's going over a certain%. 
```

```python
best_picks = loaded_data_sorted[
    (loaded_data_sorted['Predictions'] > 8.5) & 
    (loaded_data_sorted['purchase_price'] > 0.5) & 
    (loaded_data_sorted['dof_price'] > 0.5)
]
```
